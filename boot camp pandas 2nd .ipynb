```python


df['Segment'] = pd.cut(df['Profit'], bins=[-float('inf'), df['Profit'].median(), float('inf')], labels=['LowSpending', 'HighSpending'])

# Calculate average purchase quantity and total spending for each segment
segment_summary = df.groupby('Segment').agg({
    'QuantitySold': 'mean',
    'Profit': 'sum',
}).reset_index()

# Identify preferred product categories for each segment
preferred_categories = df.groupby(['Segment', 'Category']).size().unstack(fill_value=0).idxmax(axis=1).reset_index(name='PreferredCategory')

# Display results
print("Customer Segments Summary:")
print(segment_summary)

print("\nPreferred Categories for Each Segment:")
print(preferred_categories)

# Calculate average purchase quantity and total spending for each segment
segment_summary = df.groupby('Segment').agg({
    'QuantitySold': 'mean',
    'Profit': 'sum',
}).reset_index()

# Display results
print("Customer Segments Summary:")
print(segment_summary)

```
