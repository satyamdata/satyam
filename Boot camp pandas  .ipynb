```python

# Calculate total revenue for each product
total_revenue_per_product = df.groupby('ProductID')['Price'].sum().reset_index()

# Calculate average profit margin for each product
df['Profit Margin'] = (df['Profit'] / df['Price']) * 100
avg_profit_margin_per_product = df.groupby('ProductID')['Profit Margin'].mean().reset_index()

# Identify top-selling products
top_selling_products = df.groupby('ProductID')['QuantitySold'].sum().sort_values(ascending=False).reset_index()

# Identify top-selling categories
top_selling_categories = df.groupby('Category')['QuantitySold'].sum().sort_values(ascending=False).reset_index()

# Display results
print("Total Revenue per Product:")
print(total_revenue_per_product)

print("\nAverage Profit Margin per Product:")
print(avg_profit_margin_per_product)

print("\nTop Selling Products:")
print(top_selling_products.head())

print("\nTop Selling Categories:")
print(top_selling_categories.head())
```
